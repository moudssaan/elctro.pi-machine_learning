```python
def img_processing(img_path):
    img = cv2.imread(img_path,1)
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    img = cv2.resize(gray,(400,650))
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)
    gaussian = cv2.adaptiveThreshold(blurred,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,91,41)
    # Apply thresholding to make text more prominent
    _, binary_image = cv2.threshold(gaussian, 0, 10, cv2.THRESH_BINARY)
    return binary_image
```

```python
def extract_data(img):
    name_part = img[int(img.shape[0]*0.3) : int(img.shape[0]*0.5)]
    name = pytesseract.image_to_string(name_part, lang='ara').strip().replace('\n',' ')
    address_part = img[int(img.shape[0]*0.45) : int(img.shape[0]*0.7)]
    address = pytesseract.image_to_string(address_part, lang='ara').strip().replace('\n',' ')
    id_part = img[int(img.shape[0]*0.7) : int(img.shape[0]*0.8)]
    id = pytesseract.image_to_string(id_part ,lang='hin')
    return {'name':name, 'address':address,'id':id}
```

```python
processsed_img = img_processing(img_path)
print(extract_data(processsed_img))
```

```python
def main():
    img_path = input('Enter the path for the ID: ')
    processsed_img = img_processing(img_path)
    print(extract_data(processsed_img))
```
